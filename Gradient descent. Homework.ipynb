```python
import numpy as np
```

```python
N = 5  # Количество входных данных тестовой выборки

X = np.array([ #Тестовая выборка
    [0, 1],
    [1, 1],
    [1, 0],
    [-0.5, 0.5],
    [0, -0.5]
])

y = np.array([1, 1, 1, -1, -1]) #Целевые значения

theta0 = np.array([1.0, 1.0, 0.0]) #Веса

lr = 0.1
```

```python
def f(X, theta):
    theta = np.asarray(theta)
    return (X * theta[:2]).sum(axis=-1) + theta[2]

def loss(X, y, theta):
    theta = np.asarray(theta)
    norm = (theta ** 2).sum()
    deltas = y * f(X, theta)
    return 0.1 * norm + np.mean(np.maximum(0, 1 - deltas))
```

Найдем оптимумы по трем параметрам с помощью derL1 derL2 derL3

```python

def derL1(theta, x, y):
    res = 0
    for i in range(N):
        value = -y[i] * x[i][0]
        if ((1 - (y[i] * f(x[i], theta))) > 0):
            res += value
    res /= N
    res += 0.2 * theta[0]
    return res

def derL2(theta, x, y):
    res = 0
    for i in range(N):
        value = -y[i] * x[i][1]
        if (1 - y[i] * f(x[i], theta) < 0):
            value = 0
        res += value
    res /= N
    res += 0.2 * theta[1]
    return res

def derL3(theta, x, y):
    res = 0
    for i in range(N):
        value = -y[i] * 1
        if (1 - y[i] * f(x[i], theta) < 0):
            value = 0
        res += value
    res /= N
    res += 0.2 * theta[2]
    return res
```

Сама функция градиентного спуска

```python

def GD(theta, X, y):
    prev_tht = theta
    for i in range(1000):
        tht_1 = prev_tht[0] - lr * derL1(prev_tht, X, y)
        tht_2 = prev_tht[1] - lr * derL2(prev_tht, X, y)
        tht_3 = prev_tht[2] - lr * derL3(prev_tht, X, y)

        prev_tht = [tht_1, tht_2, tht_3]
    return prev_tht
```

Вычислим функцию потерь

```python
theta1 = GD(theta0, X, y)

print("Prediction:", f(X, theta1))
print("Loss:", loss(X, y, theta1))
```
